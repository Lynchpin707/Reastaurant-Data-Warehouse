```python
%sql
CREATE SCHEMA IF NOT EXISTS workspace.restaurantdb;
```

```python
%sql
CREATE VOLUME IF NOT EXISTS workspace.restaurantdb.raw_data;
```

Before ingesting the data, you should upload the raw data csv files in the volume ( Catalog -> workspace -> restaurantdb -> raw_data then click on upload to volume and select the files ).


```python
%sql
List '/Volumes/workspace/restaurantdb/raw_data'
```

```python
raw_tables_names = ["details_raw.csv", "menu_raw.csv", "employees_raw.csv", "orders_raw.csv", "weather_raw.csv"]
source_volume = "/Volumes/workspace/restaurantdb/raw_data"
def initialise_restaurantdb(tables_list, volume , schema = "workspace.restaurantdb") : 
    for table_name in tables_list:
        df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(volume+"/"+table_name)
        table_full_name = f"{schema}.{table_name.replace("_raw.csv", "")}"
        df.write.mode("overwrite").saveAsTable(table_full_name)
        print(f"successfully ingested {table_full_name}")

initialise_restaurantdb(raw_tables_names, source_volume)
```
